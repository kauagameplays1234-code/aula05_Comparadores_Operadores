```python
#operadores de comparação
print("8 é maior que 6: ", 8 > 6 )
print("8 é menor que 7: ", 8 < 7)
print(8 == 7)
print(10 >= 10)
print(10 < 10)
print(10 <= 10)
```

```python
#Atividade 01
idade = int(input("digite a primeira idade: "))
idade2 = int(input("digite a segunda idade: "))

print("é igual ? ",idade == idade2)
print("é diferente ? ",idade != idade2)
print("é maior ? ",idade > idade2)
print("é maior ou igual ? ",idade >= idade2)
print("é menor ? ",idade < idade2)
print("é menor ou igual ? ",idade <= idade2)
```

```python
#operadores lógicos / Atividade 01
idade = int(input("digite sua idade: "))
convite = input("você possui convite ? (sim/nao)")
resultado = (idade>=18) and (convite == "sim")

print(resultado)
```

```python
#Atividade 02
usuario = input("digite seu usuário: ")
senha = input("digite sua senha: ")
resultado = (usuario=="admin") and (senha=="1234")

print(resultado)
```

```python
#Atividade 03
vip = input("você é cliente vip ? (sim/nao)")
valor = float(input("digite o valor da compra: "))
resultado = (vip=="sim") or (valor>100)

print(resultado)
```
